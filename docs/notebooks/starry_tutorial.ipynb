```python
planet = starry.Planet()
```

```python
planet.map[1,0] = 0.5
planet.map[2,2] = 1.5
```

```python
planet.map
```

```python
planet.map.show()
```

```python
star = starry.Star()
```

```python
star.map.set_coeff(1,0,0.5)
```

```python
star.map.show()
```

```python
system = starry.System([star, planet])
```

```python
time = np.linspace(-1.5,1.5,1000)

system.compute(time)

flux = system.flux

flux_norm = (flux / np.median(flux))
```

```python
plt.plot(time, flux_norm)
```
